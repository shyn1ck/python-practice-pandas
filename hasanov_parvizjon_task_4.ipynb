#### –ó–∞–¥–∞—á–∞ 1. –†–∞–±–æ—Ç–∞ —Å DataFrame –≤ –±–∏–±–ª–∏–æ—Ç–µ–∫–µ pandas

1) –°–æ–∑–¥–∞–π—Ç–µ DataFrame —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å—é 5—Ö5 —Å –¥–∞–Ω–Ω—ã–º–∏ –æ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö —Å—Ç—É–¥–µ–Ω—Ç–∞—Ö –∏ –∏—Ö –æ—Ü–µ–Ω–∫–∞—Ö –ø–æ —Ä–∞–∑–Ω—ã–º –ø—Ä–µ–¥–º–µ—Ç–∞–º. –í–∫–ª—é—á–∏—Ç–µ —Å–ª–µ–¥—É—é—â–∏–µ —Å—Ç–æ–ª–±—Ü—ã: `'–ò–º—è', '–í–æ–∑—Ä–∞—Å—Ç', '–ú–∞—Ç–µ–º–∞—Ç–∏–∫–∞', '–õ–∏—Ç–µ—Ä–∞—Ç—É—Ä–∞', '–§–∏–∑–∏–∫–∞'` –∏ –≤—ã–≤–µ–¥–∏—Ç–µ –∏—Å—Ö–æ–¥–Ω—ã–π DataFrame –Ω–∞ —ç–∫—Ä–∞–Ω;

2) –ò–∑—É—á–∏—Ç–µ –≤–∞—à DataFrame: —Ä–∞—Å—Å–º–æ—Ç—Ä–∏—Ç–µ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å, –∏–º–µ—é—â–∏–µ—Å—è –∫–æ–ª–æ–Ω–∫–∏, —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö, –±–∞–∑–æ–≤—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é;

3) –ò—Å–ø–æ–ª—å–∑—É—è –º–µ—Ç–æ–¥ `loc`, –≤—ã–±–µ—Ä–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ –ø–æ –æ–¥–Ω–æ–º—É –∏–∑ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤, —Å –ø–æ—Å–ª–µ–¥—É—é—â–∏–º –≤—ã–≤–æ–¥–æ–º –Ω–∞ —ç–∫—Ä–∞–Ω; 

4) –ò—Å–ø–æ–ª—å–∑—É—è –º–µ—Ç–æ–¥ `iloc`, –≤—ã–±–µ—Ä–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ –ø–æ –ø–µ—Ä–≤–æ–º—É —Å—Ç—É–¥–µ–Ω—Ç—É, —Å –ø–æ—Å–ª–µ–¥—É—é—â–∏–º –≤—ã–≤–æ–¥–æ–º –Ω–∞ —ç–∫—Ä–∞–Ω. –ó–∞—Ç–µ–º –Ω–∞–π–¥–∏—Ç–µ –µ–≥–æ –æ—Ü–µ–Ω–∫—É –ø–æ —Ñ–∏–∑–∏–∫–µ;

5) –î–æ–±–∞–≤—å—Ç–µ –¥–≤–µ –∫–æ–ª–æ–Ω–∫–∏: `–ü–æ–ª` –∏ `–ì–æ—Ä–æ–¥`; 

6) –° –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ `loc` –∏–∑–º–µ–Ω–∏—Ç–µ –æ—Ü–µ–Ω–∫—É –ø–æ –º–∞—Ç–µ–º–∞—Ç–∏–∫–µ –¥–ª—è –æ–¥–Ω–æ–≥–æ —Å—Ç—É–¥–µ–Ω—Ç–∞;

7) –î–æ–±–∞–≤—å—Ç–µ –Ω–æ–≤—É—é —Å—Ç—Ä–æ–∫—É –≤ –≤–∞—à DataFrame;

8) –í—ã–≤–µ–¥–∏—Ç–µ –∏—Ç–æ–≥–æ–≤—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç.

```python
import pandas as pd

data = {
    '–ò–º—è': ['–ê–ª–µ–∫—Å–µ–π', '–ú–∞—Ä–∏—è', '–ò–≤–∞–Ω', '–ê–Ω–Ω–∞', '–ü–µ—Ç—Ä'],
    '–í–æ–∑—Ä–∞—Å—Ç': [20, 19, 21, 20, 22],
    '–ú–∞—Ç–µ–º–∞—Ç–∏–∫–∞': [90, 75, 88, 92, 85],
    '–õ–∏—Ç–µ—Ä–∞—Ç—É—Ä–∞': [80, 85, 78, 88, 90],
    '–§–∏–∑–∏–∫–∞': [85, 80, 82, 90, 88]
}
df = pd.DataFrame(data)
print("–ò—Å—Ö–æ–¥–Ω—ã–π DataFrame:")
print(df)

print("\n–†–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å DataFrame:", df.shape)
print("–ö–æ–ª–æ–Ω–∫–∏:", df.columns)
print("–¢–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö:\n", df.dtypes)
print("\n–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ DataFrame:")
print(df.info())

print("\n–î–∞–Ω–Ω—ã–µ –ø–æ —Å—Ç—É–¥–µ–Ω—Ç—É '–ú–∞—Ä–∏—è':")
print(df.loc[df['–ò–º—è'] == '–ú–∞—Ä–∏—è'])

print("\n–î–∞–Ω–Ω—ã–µ –ø–æ –ø–µ—Ä–≤–æ–º—É —Å—Ç—É–¥–µ–Ω—Ç—É:")
print(df.iloc[0])
print("\n–û—Ü–µ–Ω–∫–∞ –ø–æ —Ñ–∏–∑–∏–∫–µ —É –ø–µ—Ä–≤–æ–≥–æ —Å—Ç—É–¥–µ–Ω—Ç–∞:", df.iloc[0]['–§–∏–∑–∏–∫–∞'])

df['–ü–æ–ª'] = ['–ú', '–ñ', '–ú', '–ñ', '–ú']
df['–ì–æ—Ä–æ–¥'] = ['–ú–æ—Å–∫–≤–∞', '–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥', '–ö–∞–∑–∞–Ω—å', '–ù–æ–≤–æ—Å–∏–±–∏—Ä—Å–∫', '–ï–∫–∞—Ç–µ—Ä–∏–Ω–±—É—Ä–≥']
print("\nDataFrame —Å –Ω–æ–≤—ã–º–∏ –∫–æ–ª–æ–Ω–∫–∞–º–∏ –ü–æ–ª –∏ –ì–æ—Ä–æ–¥:")
print(df)

df.loc[df['–ò–º—è'] == '–ò–≤–∞–Ω', '–ú–∞—Ç–µ–º–∞—Ç–∏–∫–∞'] = 95
print("\nDataFrame –ø–æ—Å–ª–µ –∏–∑–º–µ–Ω–µ–Ω–∏—è –æ—Ü–µ–Ω–∫–∏ –ø–æ –º–∞—Ç–µ–º–∞—Ç–∏–∫–µ –¥–ª—è —Å—Ç—É–¥–µ–Ω—Ç–∞ –ò–≤–∞–Ω:")
print(df)

new_student = pd.DataFrame([{
    '–ò–º—è': '–û–ª—å–≥–∞',
    '–í–æ–∑—Ä–∞—Å—Ç': 23,
    '–ú–∞—Ç–µ–º–∞—Ç–∏–∫–∞': 89,
    '–õ–∏—Ç–µ—Ä–∞—Ç—É—Ä–∞': 91,
    '–§–∏–∑–∏–∫–∞': 87,
    '–ü–æ–ª': '–ñ',
    '–ì–æ—Ä–æ–¥': '–°–∞–º–∞—Ä–∞'
}])

df = pd.concat([df, new_student], ignore_index=True)

print("\n–ò—Ç–æ–≥–æ–≤—ã–π DataFrame:")
print(df)
```

#### –ó–∞–¥–∞—á–∞ 2. –û–±—Ä–∞–±–æ—Ç–∫–∞ –¥–æ—Ö–æ–¥–æ–≤ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤

1) –°–æ–∑–¥–∞–π—Ç–µ –æ–¥–∏–Ω DataFrame —Å –¥–æ—Ö–æ–¥–∞–º–∏ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ –∑–∞ 3 –º–µ—Å—è—Ü–∞:

    - –°–æ—Ç—Ä—É–¥–Ω–∏–∫ 1: 5000, 5500, 6000
    - –°–æ—Ç—Ä—É–¥–Ω–∏–∫ 2: 6000, 6500, 7000
    - –°–æ—Ç—Ä—É–¥–Ω–∏–∫ 3: 4500, 4800, 5100
    - –°–æ—Ç—Ä—É–¥–Ω–∏–∫ 4: 7000, 7500, 8000
    - –°–æ—Ç—Ä—É–¥–Ω–∏–∫ 5: 5500, 5800, 6100

2) –ü—Ä–µ–æ–±—Ä–∞–∑—É–π—Ç–µ Series –≤ DataFrame, –≥–¥–µ —Å—Ç–æ–ª–±—Ü—ã –±—É–¥—É—Ç –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—Ç—å –º–µ—Å—è—Ü—ã (–Ø–Ω–≤–∞—Ä—å, –§–µ–≤—Ä–∞–ª—å, –ú–∞—Ä—Ç);

3) –°–æ–∑–¥–∞–π—Ç–µ –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü `–°—É–º–º–∞` –∏ `–°—Ä–µ–¥–Ω—è—è`, –∫–æ—Ç–æ—Ä—ã–π –±—É–¥–µ—Ç —Å–æ–¥–µ—Ä–∂–∞—Ç—å —Å—É–º–º—É –¥–æ—Ö–æ–¥–æ–≤ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–∞ –∑–∞ 3 –º–µ—Å—è—Ü–∞;

4) –í—ã–≤–µ–¥–∏—Ç–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–µ —Å –Ω–∞–∏–±–æ–ª—å—à–µ–π —Å—É–º–º–æ–π –¥–æ—Ö–æ–¥–∞ –∑–∞ 3 –º–µ—Å—è—Ü–∞.

```python
# –†–µ—à–µ–Ω–∏–µ
import pandas as pd

data = {
    '–°–æ—Ç—Ä—É–¥–Ω–∏–∫': ['–°–æ—Ç—Ä—É–¥–Ω–∏–∫ 1', '–°–æ—Ç—Ä—É–¥–Ω–∏–∫ 2', '–°–æ—Ç—Ä—É–¥–Ω–∏–∫ 3', '–°–æ—Ç—Ä—É–¥–Ω–∏–∫ 4', '–°–æ—Ç—Ä—É–¥–Ω–∏–∫ 5'],
    '–Ø–Ω–≤–∞—Ä—å': [5000, 6000, 4500, 7000, 5500],
    '–§–µ–≤—Ä–∞–ª—å': [5500, 6500, 4800, 7500, 5800],
    '–ú–∞—Ä—Ç': [6000, 7000, 5100, 8000, 6100]
}
df_income = pd.DataFrame(data)
df_income['–°—É–º–º–∞'] = df_income[['–Ø–Ω–≤–∞—Ä—å', '–§–µ–≤—Ä–∞–ª—å', '–ú–∞—Ä—Ç']].sum(axis=1)
df_income['–°—Ä–µ–¥–Ω—è—è'] = df_income[['–Ø–Ω–≤–∞—Ä—å', '–§–µ–≤—Ä–∞–ª—å', '–ú–∞—Ä—Ç']].mean(axis=1)

highest_income_employee = df_income.loc[df_income['–°—É–º–º–∞'].idxmax()]
print("\n–°–æ—Ç—Ä—É–¥–Ω–∏–∫ —Å –Ω–∞–∏–±–æ–ª—å—à–µ–π —Å—É–º–º–æ–π –¥–æ—Ö–æ–¥–∞ –∑–∞ 3 –º–µ—Å—è—Ü–∞:")
print(highest_income_employee)
```

#### –ó–∞–¥–∞—á–∞ 3. –°–æ–∑–¥–∞–Ω–∏–µ –ø—Ä–æ–≤–µ—Ä–æ–∫ –¥–ª—è –≤—ã—è–≤–ª–µ–Ω–∏—è –Ω–µ—Ö–∞—Ä–∞–∫—Ç–µ—Ä–Ω—ã—Ö –∫–µ–π—Å–æ–≤.

1) –°–æ–∑–¥–∞—Ç—å `DataFrame` 10x10 c –§–ò–û —É—á–µ–Ω–∏–∫–æ–≤ –∏ –∏—Ö –æ—Ü–µ–Ω–æ–∫ –ø–æ –¥–∏—Å–ø–∏—Ü–ª–∏–Ω–∞–º. –í –¥–∞–Ω–Ω–æ–º `DataFrame` –¥–æ–ª–∂–Ω—ã –ø—Ä–∏—Å—É—Ç—Å—Ç–≤–æ–≤–∞—Ç—å –¥—É–±–ª–∏–∫–∞—Ç—ã –∏ –Ω—É–ª–µ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è `(np.nan)`. 

2) –í—ã–≤–µ–¥–∏—Ç–µ –∫—Ä–∞—Ç–∫—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é DataFrame: –∫–æ–ª-–≤–æ —Å—Ç—Ä–æ–∫ –∏ –∫–æ–ª–æ–Ω–æ–∫, –Ω–∞–∑–≤–∞–Ω–∏–µ –∫–æ–ª–æ–Ω–æ–∫, –∫—Ä–∞—Ç–∫–∞—è —Å–≤–æ–¥–∫–∞ –ø–æ —Ç–∞–±–ª–∏—Ü–µ + —Ç–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö;

3)  –í—ã–≤–µ–¥–∏—Ç–µ –∫—Ä–æ—Å—Å-—á–µ–∫ (–ø—Ä–æ–≤–µ—Ä–∫—É), –∫–æ—Ç–æ—Ä–∞—è –ø–æ–∑–≤–æ–ª–∏—Ç –≤–∞–º –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å:
    - –î—É–±–ª–∏–∫–∞—Ç—ã (–≤—ã–≤–æ–¥ –æ—à–∏–±–∫–∏ –∏ —Å–∞–º–∏—Ö –¥—É–±–ª–∏—Ä—É—é—â–∏—Ö—Å—è —Å—Ç—Ä–æ–∫);
    - –ù—É–ª–µ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏–µ (–≤—ã–≤–æ–¥ –æ—à–∏–±–∫–∏ –∏ —Å–∞–º–∏—Ö –Ω—É–ª–µ–≤—ã—Ö —Å—Ç—Ä–æ–∫);
    - –ü—Ä–æ–≤–µ—Ä–∫—É –Ω–∞ –ø—É—Å—Ç–æ—Ç—É –≤–∞—à–µ–≥–æ `DataFrame` (–æ—à–∏–±–∫–∞).

```python
# –†–µ—à–µ–Ω–∏–µ
import pandas as pd
import numpy as np

data = {
    '–§–ò–û': ['–ò–≤–∞–Ω–æ–≤ –ò–≤–∞–Ω', '–ü–µ—Ç—Ä–æ–≤ –ü–µ—Ç—Ä', '–°–∏–¥–æ—Ä–æ–≤ –°–∏–¥–æ—Ä', '–ò–≤–∞–Ω–æ–≤ –ò–≤–∞–Ω', 
            '–ü–µ—Ç—Ä–æ–≤ –ü–µ—Ç—Ä', '–°–∏–¥–æ—Ä–æ–≤ –°–∏–¥–æ—Ä', '–ö—É–∑–Ω–µ—Ü–æ–≤ –ö—É–∑—å–º–∞', '–ü–µ—Ç—Ä–æ–≤ –ü–µ—Ç—Ä', 
            '–°–∏–¥–æ—Ä–æ–≤ –°–∏–¥–æ—Ä', '–ò–≤–∞–Ω–æ–≤ –ò–≤–∞–Ω'],
    '–ú–∞—Ç–µ–º–∞—Ç–∏–∫–∞': [5, 4, 3, 5, np.nan, 4, 3, 5, np.nan, 2],
    '–§–∏–∑–∏–∫–∞': [4, 5, np.nan, 4, 3, 5, 4, np.nan, 2, 5],
    '–•–∏–º–∏—è': [3, 3, 4, 2, 5, np.nan, 4, 5, 4, 3],
    '–õ–∏—Ç–µ—Ä–∞—Ç—É—Ä–∞': [4, 5, 4, 3, np.nan, 2, 4, 3, 5, 5],
    '–ò—Å—Ç–æ—Ä–∏—è': [5, np.nan, 4, 4, 5, 3, 4, 5, 5, 3],
    '–ì–µ–æ–≥—Ä–∞—Ñ–∏—è': [np.nan, 4, 3, 5, 4, 2, 5, 5, 4, 4],
    '–ò–Ω–æ—Å—Ç—Ä–∞–Ω–Ω—ã–π —è–∑—ã–∫': [5, 3, 4, 5, 5, np.nan, 3, 4, 5, 4],
    '–§–∏–∑–∏—á–µ—Å–∫–∞—è –∫—É–ª—å—Ç—É—Ä–∞': [4, 5, np.nan, 3, 2, 4, 3, 5, 4, 5],
    '–ú—É–∑—ã–∫–∞': [3, 4, 5, 5, 4, 5, 3, 2, 5, 3]
}

df_students = pd.DataFrame(data)

print("–ö—Ä–∞—Ç–∫–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ DataFrame:")
print(df_students.info())

duplicates = df_students[df_students.duplicated(keep=False)]
if not duplicates.empty:
    print("\n–û—à–∏–±–∫–∞: –ù–∞–π–¥–µ–Ω—ã –¥—É–±–ª–∏–∫–∞—Ç—ã:")
    print(duplicates)
else:
    print("\n–î—É–±–ª–∏–∫–∞—Ç–æ–≤ –Ω–µ –Ω–∞–π–¥–µ–Ω–æ.")

null_values = df_students[df_students.isnull().any(axis=1)]
if not null_values.empty:
    print("\n–û—à–∏–±–∫–∞: –ù–∞–π–¥–µ–Ω—ã —Å—Ç—Ä–æ–∫–∏ —Å –Ω—É–ª–µ–≤—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏:")
    print(null_values)
else:
    print("\n–ù—É–ª–µ–≤—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –Ω–µ –Ω–∞–π–¥–µ–Ω–æ.")

if df_students.empty:
    print("\n–û—à–∏–±–∫–∞: DataFrame –ø—É—Å—Ç.")
else:
    print("\nDataFrame –Ω–µ –ø—É—Å—Ç.")
```
